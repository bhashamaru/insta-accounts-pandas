```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
instalist = pd.read_csv("List of most-followed Instagram accounts.csv", index_col=0)
instalist = instalist.rename(columns={"Followers(millions)[2]":"Followers", "Profession/Activity":"Profession"})
instalist
```

```python
#Dropping last row which is not part of the sample data of instagram handle and their followers
instalist = instalist.drop(50)
```

```python
#setting index column as Rank instead of redundant unnamed col from csv file
instalist = instalist.set_index("Rank") 
```

```python
#checking memory usage before converting columns to appropriate datatypes
instalist.info()
```

Converting columns to appropriate datatypes

```python
instalist.nunique()
```

```python
instalist["Country/Continent"] = instalist["Country/Continent"].astype("category")
instalist["Followers"] = instalist["Followers"].astype("float")
```

```python
instalist.info()
```

```python
instalist.head()
```

```python
#Top 10 most followed insta accounts
instalist.nlargest(10,columns="Followers")
```

```python
top10owners = instalist.nlargest(10,columns="Followers").groupby("Owner")["Followers"].sum().sort_values(ascending=False)
top10owners
```

```python
# Plot bar chart
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
bars = top10owners.plot(kind='bar', color='skyblue')
plt.title('Bar Chart of Insta followers grouped by their account owners')
plt.xlabel('Instagram Account Owners')
plt.ylabel('Instagram Followers')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Annotate bars with their values
for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
             '{:.0f}'.format(bar.get_height()), 
             ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

Exploring Profession data

```python
instalist["Profession"].value_counts()
```

Finding the profession with most number of insta followers by using string methods

```python
#Actual count of insta owners who are musicians
professionMusician = instalist["Profession"].str.lower().str.contains("musician")
instalist[professionMusician]["Profession"].count()
```

```python
#Actual count of insta owners who are actors/actresses
professionActors = instalist["Profession"].str.lower().str.contains("actr")
instalist[professionActors]["Profession"].count()
```

```python
#Actual count of insta owners who are footballers
professionFootball = instalist["Profession"].str.lower().str.contains("football")
instalist[professionFootball]["Profession"].count()
```

Hence, I can conclude from this data using string methods that the profession with most insta followers is Musician.

Exploring Country/Continent data

```python
unique_counts_country = instalist["Country/Continent"].value_counts().nlargest(n=5)
unique_counts_country
```

```python
# Plot bar chart
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
bars = unique_counts_country.plot(kind='bar', color='skyblue')
plt.title('Bar Chart of Country/Continent vs count of insta accounts')
plt.xlabel('Country/Continent')
plt.ylabel('Count of top insta accounts')
plt.xticks(rotation=45)  #Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Annotate bars with their values
for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
             '{:.0f}'.format(bar.get_height()), 
             ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

I can clearly conclude from this country/continent data that United States has the most number of insta accounts with most number of followers.
